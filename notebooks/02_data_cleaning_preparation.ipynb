### GDP in dolars and in % growth - 1961-2021

```python
gdp_columns = []
for column in wb.columns:
    if 'gdp_per_capita' in column:
        gdp_columns.append(column)
    else:
        pass
    
gdp_columns.append('time')
gdp_columns
```

```python
wb_gdp = wb[gdp_columns]
wb_gdp = wb_gdp[1:].reset_index(drop=True) # eliminating 1960 because there is no information for GDP growth
wb_gdp.isna().sum()
```

```python
wb_gdp.head()
```

```python
# I can calculate the Argentinian value
wb_gdp['ARG_gdp_per_capita_us$'][0] = (wb_gdp['ARG_gdp_per_capita_us$'][1]*100)/(100+(wb_gdp['ARG_gdp_per_capita_%_growth'][1]))
wb_gdp['ARG_gdp_per_capita_us$'][0]
```

```python
wb_gdp['CAN_gdp_per_capita_%_growth'].head(60)
```

```python
# Those can also be calculated

def growth_calculation(gdp_year_1, gdp_year_2):
    gdp_percentage_growth = ((gdp_year_2 - gdp_year_1)/gdp_year_1)*100
    
    return gdp_percentage_growth

for i in range (9,37):
    wb_gdp['CAN_gdp_per_capita_%_growth'][i] = growth_calculation(wb_gdp['CAN_gdp_per_capita_us$'][i-1],wb_gdp['CAN_gdp_per_capita_us$'][i])
    
wb_gdp['CAN_gdp_per_capita_%_growth'].head(40)
```

```python
wb_gdp['IDN_gdp_per_capita_us$'].head(10)
```

```python
wb_gdp['IDN_gdp_per_capita_%_growth'].head(10)
```

```python
# Those can also be calculated

def gdp_calculation(gdp_year_2, growth_year_2):
    gdp_year1 = (gdp_year_2*100)/(100+growth_year_2)
    
    return gdp_year1

for i in reversed(range(6)):
      wb_gdp['IDN_gdp_per_capita_us$'][i] = gdp_calculation(wb_gdp['IDN_gdp_per_capita_us$'][i+1],wb_gdp['IDN_gdp_per_capita_%_growth'][i+1])
    
wb_gdp['IDN_gdp_per_capita_us$'].head(10)
```

```python
wb_gdp['DEU_gdp_per_capita_us$'].head(30)
```

```python
wb_gdp['RUS_gdp_per_capita_us$'].head(30)
```

```python
wb_gdp['SAU_gdp_per_capita_us$'].head(10)
```

```python
# I do not want to lose this data so I will fill the NaN in growth with the mean and calculate the GDP

mean_DEU = wb_gdp['DEU_gdp_per_capita_%_growth'].mean()
wb_gdp['DEU_gdp_per_capita_%_growth'] = wb_gdp['DEU_gdp_per_capita_%_growth'].fillna(mean_DEU)
mean_RUS = wb_gdp['RUS_gdp_per_capita_%_growth'].mean()
wb_gdp['RUS_gdp_per_capita_%_growth'] = wb_gdp['RUS_gdp_per_capita_%_growth'].fillna(mean_RUS)
mean_SAU = wb_gdp['SAU_gdp_per_capita_%_growth'].mean()
wb_gdp['SAU_gdp_per_capita_%_growth'] = wb_gdp['SAU_gdp_per_capita_%_growth'].fillna(mean_SAU)
```

```python
for i in reversed(range(9)):
      wb_gdp['DEU_gdp_per_capita_us$'][i] = gdp_calculation(wb_gdp['DEU_gdp_per_capita_us$'][i+1],
                                                            wb_gdp['DEU_gdp_per_capita_%_growth'][i+1])

        
for i in reversed(range(27)):
      wb_gdp['RUS_gdp_per_capita_us$'][i] = gdp_calculation(wb_gdp['RUS_gdp_per_capita_us$'][i+1],
                                                            wb_gdp['RUS_gdp_per_capita_%_growth'][i+1])
        
for i in reversed(range(7)):
      wb_gdp['SAU_gdp_per_capita_us$'][i] = gdp_calculation(wb_gdp['SAU_gdp_per_capita_us$'][i+1],
                                                            wb_gdp['SAU_gdp_per_capita_%_growth'][i+1])
```

```python
wb_gdp.isna().sum()
```

```python
# I will drop the subjects which do not belong to the list of most relevant for global crisis
subject_to_drop = ['Population',
                   'General government revenue',
                   'Gross domestic product corresponding to fiscal year, current prices',
                   'Current account balance',
                   'Gross domestic product, deflator',
                   'Gross domestic product based on purchasing-power-parity (PPP) share of world total',
                   'Implied PPP conversion rate',
                   'Inflation, end of period consumer prices',
                   'General government total expenditure',
                   'General government net lending/borrowing',
                   'General government primary net lending/borrowing',
                   'Total investment',
                   'Gross national savings',
                   'Volume of Imports of goods',
                   'Volume of exports of goods',
                   'General government structural balance']
```

```python
for column in imf.columns:
    for i in range(len(imf)):
        try:
            if imf[column].isna()[i]==True and imf.iloc[i]['Subject Descriptor'] in subject_to_drop:
                imf = imf.drop(index=i, axis=1)
        except:
            pass
imf = imf.reset_index(drop=True)
imf.isna().sum()
```

```python
# For the remaining NaNs I will use the means

imf['Estimates Start After'].fillna('No information')
        
years_to_2019 = ['1980', '1981','1982', '1983', '1984', '1985', '1986', '1987', '1988', '1989', '1990','1991', '1992',
               '1993', '1994', '1995', '1996', '1997', '1998', '1999','2000', '2001', '2002', '2003', '2004', '2005',
               '2006', '2007', '2008','2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017','2018',
               '2019']
# Choosing until 2019 to avoid noise after covid and mostly estimated values

for column in imf.columns:
    for i in range(len(imf)):
        try:
            if imf[column].isna()[i]==True:
                mean = imf.loc[i][years_to_2019].mean(skipna=True)
                imf.loc[i][column] = mean
        except:
            pass

imf = imf.reset_index(drop=True)
imf.isna().sum()
```
